```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sqlite3
 
conn = sqlite3.connect("MAI.db")
```

```python
#conn.row_factory = sqlite3.Row
cursor = conn.cursor()
 
sql = "SELECT * FROM tracking WHERE time=?"
cursor.execute(sql, [("9.10")])
print(cursor.fetchall()) # or use fetchone()
 
```

```python
print("Here's a listing of all the records in the table:")
for row in cursor.execute("SELECT rowid, * FROM tracking ORDER BY time"):
    print(row)
 
print("Results from a LIKE query:")
sql = "SELECT * FROM tracking WHERE time LIKE '9.0%'"
cursor.execute(sql)
 
print(cursor.fetchall())
```

```python
print("Here's a listing of GPS-coordinats of the records in the table tracking:")
for row in cursor.execute("SELECT  longitude,latitude FROM tracking ORDER BY time"):
    print(row)
```
